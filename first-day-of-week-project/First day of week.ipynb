```python
import jupyter_black
import pandas as pd
import matplotlib.pyplot as plt

jupyter_black.load()
```

```python
df = pd.read_csv("first-day-of-week.csv")
df
```

## Removing the paper section as it not needed

```python
df = df.drop("paper", axis=1)
```

## Group by country starting day of the week

```python
day_counts = df.groupby("first_day").size()
day_counts
```

```python
plt.bar(day_counts.index, day_counts.values, color="skyblue")
plt.xlabel("First day")
plt.ylabel("Territory")
plt.title("Country first day")
plt.show()
```
